```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
```

```python
plt.style.use('fivethirtyeight')
```

```python
data=sns.load_dataset('tips')
data.head()
```

## **Scater plot between total_bill and tip.**

```python
# plot here
sns.relplot(x='total_bill',y='tip',data=data)
```

**The hue operator**

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',data=data)
```

**The style parameter**

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style='sex',data=data)
```

**The size parameter**

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style='sex',size='size',sizes=(10,100),data=data)
```

**Using the scatterplot function**

```python
sns.scatterplot(x='total_bill',y='tip',hue='smoker',style='sex',size='size',sizes=(10,100),data=data)
```

```python
sns.heatmap(x,cbar=False,linewidths=0.5,annot=True)
```

# **DISTPLOTS**

```python
 #DISPLOT ARE USED TO DO UNIVARIANT ANALYSIS ON A CONTINOUS /NUMERICAL data #

# Distplot combine 3 thing in one grap :
#1. Histogram 2. Kernel Density Enstimation(PDF)3. Rugplot
```

```python
titanic= sns.load_dataset('titanic')
titanic.head()
```

```python
titanic['age'].fillna(titanic['age'].mean(),inplace=True)
```

PLOTING A DISTPLOT

---



```python
sns.displot(titanic['age'])
```

PLOTING ONLY HISTOGRAM

---



```python
sns.distplot(titanic['age'],bins=10,kde=False)
```

PLOTING ONLY THE KDE

```python
sns.distplot(titanic['age'],hist= False)
```

PLTOING ONLY THE RUGPLOT

---



```python
sns.distplot(titanic['age'],hist= False, kde= False, rug= True)
```

HOW TO USE THE DISTPLOT

---



```python
sns.distplot(titanic[titanic['survived']==1]['age'])
sns.distplot(titanic[titanic['survived']==0]['age'])
```
